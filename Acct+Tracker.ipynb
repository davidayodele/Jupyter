```python
print('Importing libraries...')
# =========== importing libraries ===========
import pandas as pd
import plotly
from plotly.offline import plot
import plotly.graph_objs as go
from plotly.graph_objs import Scatter, Layout
import plotly.figure_factory as ff
from pandas import ExcelWriter
from pandas import ExcelFile
# =========== begin algs ===========
print('Done')
plotly.offline.init_notebook_mode(connected=True)  # initializes offline plotting
# ============= Excel import ================
df = pd.read_excel("cpic_now_acct.xlsx")
df = df.fillna('')

def color_cols(x): # color function
    # copy df to new - original data is not changed
    df = x.copy()
    # select all values to default value - no color (necessary for some reason)
    df.loc[:,:] = 'background-color: none'
    # overwrite default values w/ desired colors      
    df[['credit']] = 'background-color: #e5f9de'
    df[['debit']] = 'background-color: #f9dfde'    
    df.loc[35:36,1:3] = 'font-weight: bold; background-color: none'
    df.loc[36,3:4] = 'font-weight: bold'
    df.loc[27:29,:] = 'background-color: none'
    df.loc[29,1:2] = 'font-weight: bold'
    # return colored df
    return df    

# =========== Create plot ===================
print ('Building plot...')
#df = df.style.applymap(lambda v: 'background-color: %s' % 'green' if v=='col' else '')
# df_list1 = df['date'].tolist()  # extracts column as python list
# df_list2 = df['balance'].tolist()  # extracts column as list
df_list1 = df.loc[0:26,'date'].tolist()  # extracts column as python list
df_list2 = df.loc[0:26,'balance'].tolist()  # extracts column as python list

df_list3 = df.loc[31:34,'date'].tolist()  # extracts column as python list
df_list4 = df.loc[31:34,'details'].tolist()  # extracts column as python list

# Create traces
trace1 = go.Scatter(
    x = df_list1,
    y = df_list2,
    mode = 'lines+markers',
    name = 'Desert Schools',
    line = dict(
        color = ('#9c43a8'),
        width = 2)
)

trace2 = go.Scatter(
    x = df_list3,
    y = df_list4,
    mode = 'lines+markers',
    name = 'Paypal',
    line = dict(
        color = ('#f47d42'),
        width = 2)
)

data = [trace1, trace2]

layout = dict(title = 'CPIC NOW Balance History',
              yaxis = dict(title= 'Balance ($)', zeroline = True),
              xaxis = dict(title= 'Date-time', zeroline = True)
             )

fig = dict(data=data, layout=layout)

print ('Done')
df = df.style.apply(color_cols, axis=None)  # applies color function to dataframe

plotly.offline.iplot(fig, filename='cpic_acct_data1.html')
df #  displays dataframe
```
